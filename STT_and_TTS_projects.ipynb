 Whisper adalah library yang menyediakan akses ke model Whisper dari OpenAI untuk melakukan transkripsi suara.

```python
import whisper
```

## Load model

Model Whisper yang akan digunakan untuk transkripsi. Dapat memilih berbagai ukuran model seperti "tiny", "base", "small", dll. Model yang lebih besar akan memberikan hasil yang lebih akurat, tetapi membutuhkan waktu lebih lama dan lebih banyak sumber daya.

```python
# Load the model (options: tiny, base, small, medium, large)
model = whisper.load_model("tiny")
```

## Audio transcribe

Fungsi ini digunakan untuk melakukan transkripsi suara dari file audio yang diberikan (dalam format .wav, .mp3, dll). Fungsi ini akan mengubah suara dalam file menjadi teks yang dapat dibaca.



```python
# Transcribe audio
result = model.transcribe("/content/How to Talk About Books - 2 Minute English Mini Podcast.wav")
```

## Result

Mengambil hasil transkripsi yang telah diproses oleh model Whisper dan menampilkannya dalam bentuk teks. Variabel result berisi informasi lebih lanjut tentang transkripsi (misalnya, waktu atau deteksi bahasa), tetapi di sini kita hanya fokus pada teksnya.

```python
# Print the text
print("Transcription:", result["text"])
```

#  gTTS (Google Text-to-Speech)

## Library

```python
!pip install gTTS
```

```python
from gtts import gTTS
from IPython.display import Audio
```

## Convert Text to Speech

Fungsi gTTS() digunakan untuk mengubah teks yang diberikan menjadi suara. Parameter lang='en' menentukan bahwa suara yang dihasilkan adalah dalam bahasa Inggris.

```python
tts = gTTS(result["text"], lang='en')
```

## Saving Sound to File

Setelah teks diubah menjadi suara, kita menyimpannya ke dalam file .mp3 untuk bisa diputar atau digunakan lagi.

```python
tts.save("output.mp3")
```

## Play Sound

Digunakan untuk memutar file audio yang baru saja disimpan (output.mp3) secara otomatis di dalam notebook. Jika dijalankan di Google Colab, suara akan langsung diputar tanpa perlu aplikasi eksternal.

```python
Audio("output.mp3", autoplay=True)
```
